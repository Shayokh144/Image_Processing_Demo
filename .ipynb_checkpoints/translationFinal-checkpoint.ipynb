```python
# -*- coding: utf-8 -*-
"""
Created on Sun Oct 13 20:13:44 2019

@author: Asif
"""
from skimage import io
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
import numpy as np
import warnings
warnings.filterwarnings("ignore")
class ImageTranslation:
    inputImageName = ''
    transformationMatrix = np.int32([[1,0,-10],[0,1,-50]])
    inputImage = np.empty([100, 100, 3], dtype= int)
    transformedImage = inputImage
    croppedImage = inputImage
    inputImageHeight = 0
    inputImageWidth = 0
    def __init__(self, inputImageName, transformationMatrix):
        self.inputImageName = inputImageName
        self.transformationMatrix = transformationMatrix
        self.readImage()
    def readImage(self):
        self.inputImage = io.imread(self.inputImageName,as_gray = True)
        self.inputImageHeight = self.inputImage.shape[0]
        self.inputImageWidth = self.inputImage.shape[1]
        self.showImage(self.inputImageName)
        return
    def doTransformation(self):
        transFormationMatrix = self.transformationMatrix
        newH, newW = self.getHeightWidthOFTransformedImage(transFormationMatrix, self.inputImage.shape)
        transformedImage = np.empty([self.inputImageHeight, self.inputImageWidth, 3], dtype= int)
        transformedImage.fill(0)
        io.imsave('transformedImage.png', transformedImage)
        
        for x in range(0, self.inputImageHeight):
            for y in range(0, self.inputImageWidth):
                originalPoint = np.int32([x,y,1])
                newPoint = transFormationMatrix.dot(originalPoint)
                newx = newPoint[0]
                newy = newPoint[1] 
                if newx < self.inputImageHeight and newy < self.inputImageWidth and newx >=0 and newy >= 0:
                    transformedImage[newx][newy] =self.inputImage[x][y]
        self.transformedImage = transformedImage
        io.imsave('transformedImage.png', transformedImage)
        self.showImage('transformedImage.png')
        return
    
    def doCropping(self):
        transFormationMatrix = self.transformationMatrix
        newH, newW = self.getHeightWidthOFTransformedImage(transFormationMatrix, self.inputImage.shape)
        transformedImage2 = np.empty([newH, newW, 3], dtype= int)
        transformedImage2.fill(0)
        io.imsave('croppedImage.png', transformedImage2)
        for x in range(0, newH ):
            for y in range(0,newW):
                originalPoint = np.int32([x,y,1])
                newPoint = transFormationMatrix.dot(originalPoint)
                if transFormationMatrix[0][2] < 0 :
                    newx = newPoint[0] + -transFormationMatrix[0][2]
                else:
                    newx = newPoint[0]
                if transFormationMatrix[1][2] < 0 :
                    newy = newPoint[1] + -transFormationMatrix[1][2]
                else:
                    newy = newPoint[1]
                
                if newx < self.inputImageHeight and newy < self.inputImageWidth and newx >=0 and newy >= 0:
                    transformedImage2[x][y] = self.transformedImage[newx][newy]
        io.imsave('croppedImage.png', transformedImage2)
        self.showImage('croppedImage.png')
        return
    
    def getHeightWidthOFTransformedImage(self,transFormationMatrix,oldShape):
        newH = oldShape[0] - abs(transFormationMatrix[0][2])
        newW = oldShape[1] - abs(transFormationMatrix[1][2])
        return newH, newW
    
    def showImage(self, src):
        img=mpimg.imread(src)
        plt.title(src)
        imgplot = plt.imshow(img)
        plt.show()
        
if __name__ == '__main__':
    inputImageName = "input.jpg"
    transFormationMatrix = np.int32([[1,0,-10],[0,1,-50]])
    imageTranslationObj = ImageTranslation(inputImageName, transFormationMatrix)
    imageTranslationObj.doTransformation()
    imageTranslationObj.doCropping()
```
